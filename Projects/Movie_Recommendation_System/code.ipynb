# Creating movie Recommendation

```python
df.head()
```

```python
moviemat = df.pivot_table(index="user_id", columns="title", values= 'rating')
```

```python
moviemat
```

```python
ratings.sort_values('num of rating', ascending=False)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
starwars_user_ratings
```

```python
#correlation score
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars
```

```python
corr_starwars.dropna(inplace= True)
corr_starwars
```

```python
corr_starwars.sort_values('Correlation', ascending=False)
```

```python
ratings
```

```python
corr_starwars = corr_starwars.join(ratings['num of rating'])
corr_starwars
```

```python
corr_starwars[corr_starwars['num of rating']>100].sort_values('Correlation', ascending=False)
```

# Predict Function

```python
def predict_movies(movie_name):
    movie_user_rating = moviemat[movie_name]
    similar_to_starwars = moviemat.corrwith(movie_user_rating)
    
    corr_movie = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    
    corr_movie = corr_movie.join(ratings['num of rating'])
    predictions= corr_movie[corr_movie['num of rating']>100].sort_values('Correlation', ascending=False)
    
    return predictions
```

```python
predictions = predict_movies("Titanic (1997)")
```
