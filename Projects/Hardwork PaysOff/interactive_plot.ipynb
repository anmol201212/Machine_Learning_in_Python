```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = pd.read_csv("./Data/Linear_X_Train.csv").values
Y = pd.read_csv("./Data/Linear_Y_Train.csv").values
```

```python
theta = np.load("./Data/ThetaList.npy")
T0 = theta[:,0]
T1 = theta[:,1]
```

```python
plt.ion()
for i in range(0,50,3):
    y_ = T1[i]*X + T0
    # Points
    plt.scatter(X,Y)
    # line
    plt.plot(X,y_,'red')
    plt.draw()
    plt.pause(1)
    plt.clf()
```
