```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('D:\Projects\TCE\TCE\Movie Review Classification\Dataset\Train.csv')
df
```

```python
X = df['review']
Y = df['label']
Y = Y.map({'pos':1,'neg':0})
```

```python
# cleaning the data 
import clean_text as ct 
x_clean = [ct.getStemmedReview(i) for i in X]
cv = CountVectorizer()
x_vec = cv.fit_transform(x_clean).toarray()
print(x_vec)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
mnb = MultinomialNB()
mnb.fit(x_vec,Y)
```

```python
df_test = pd.read_csv('D:\Projects\TCE\TCE\Movie Review Classification\Dataset\Test.csv')
X_test = df_test['review']
x_test_clean = [ct.getStemmedReview(i) for i in X_test]
cv = CountVectorizer()
x_test_vec = cv.fit_transform(x_test_clean).toarray()
print(x_test_vec)
```

```python
mnb.score(x_vec,Y)
```
