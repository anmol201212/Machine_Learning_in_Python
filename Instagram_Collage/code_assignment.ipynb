```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
center = cv2.imread('D:/Machine Learning/Project_1/Image/center.jpeg')
center = cv2.cvtColor(center,cv2.COLOR_BGR2RGB)

bottom_left = cv2.imread('D:/Machine Learning/Project_1/Image/bottom_left.jpg')
bottom_left = cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)

bottom_right = cv2.imread('D:/Machine Learning/Project_1/Image/bottom_right.jpg')
bottom_right = cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)

top_right = cv2.imread('D:/Machine Learning/Project_1/Image/top_right.jpg')
top_right = cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)

top_left = cv2.imread('D:/Machine Learning/Project_1/Image/top_left.jpg')
top_left = cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(center)
plt.show()
plt.imshow(top_right)
plt.show()
plt.imshow(top_left)
plt.show()
plt.imshow(bottom_left)
plt.show()
plt.imshow(bottom_right)
plt.show()
```

```python
center = cv2.resize(center,(100,100))
top_left = cv2.resize(top_left,(200,200))
top_right = cv2.resize(top_right,(200,200))
bottom_left = cv2.resize(bottom_left,(200,200))
bottom_right = cv2.resize(bottom_right,(200,200))
```

```python
print(center.shape)
print(top_left.shape)
print(top_right.shape)
print(bottom_left.shape)
print(bottom_right.shape)
```

```python
def create_black_canvas(height,width):
    image = np.zeros((height,width,3),np.uint8)
    return image
```

```python
canvas = create_black_canvas(430,430)
center_canvas = create_black_canvas(120,120)
plt.imshow(canvas)
plt.show()
plt.imshow(center_canvas)
plt.show()
```

```python
print(canvas.shape)
print(center_canvas.shape)
```

```python
center_canvas[10:110,10:110] = center
plt.imshow(center_canvas)
plt.show()
```

```python
canvas[10:210, 10:210] = top_left
canvas[220:420, 220:420] = bottom_left
canvas[10:210, 220:420] = top_right
canvas[220:420, 10:210] = bottom_right
canvas[155:275, 155:275] = center_canvas
plt.imshow(canvas)
plt.show()
```

```python
canvas = cv2.cvtColor(canvas,cv2.COLOR_RGB2BGR)

cv2.imwrite('sample_img.jpeg',canvas)
```

```python
r_channel = canvas[:,:,2]
g_channel = canvas[:,:,1]
b_channel = canvas[:,:,0]

r_channel = r_channel.reshape((430*430,1))
g_channel = g_channel.reshape((430*430,1))
b_channel = b_channel.reshape((430*430,1))

img = np.hstack((r_channel,g_channel,b_channel))
```

```python
print(img.shape)

df = pd.DataFrame(img,columns=['r','g','b'])

df.to_csv("Final.csv",index=False)
```
