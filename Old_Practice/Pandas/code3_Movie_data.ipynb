Task: make a scatter plot between length if the moview title and freq of occurence

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_metadata.csv")
```

```python
df
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
# print(titles)
scores = list(df.get('imdb_score'))
# print(scores)
```

```python
freq_title = {}

for t in titles:
    length = len(t)
    try:
        freq_title[length] +=1
    except:
        freq_title[length] =1

print(freq_title)
```

```python
X = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Movie Title len")
plt.ylabel("No. of Movies")
plt.show()
```
