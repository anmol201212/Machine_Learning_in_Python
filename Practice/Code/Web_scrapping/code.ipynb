```python
from urllib.request import urlopen
from http.client import IncompleteRead

url_link = "https://bidplus.gem.gov.in/all-bids"

try:
    url_data = urlopen(url_link)
    print(type(url_data))
    data_html = url_data.read()
    print(data_html)
except IncompleteRead as e:
    # Handle the incomplete read error
    print("IncompleteRead error occurred:", e)
```

```python
from bs4 import BeautifulSoup as soup 
```

```python
web_soup = soup(data_html,'html.parser')
print(web_soup)
```

```python
cards = web_soup.findAll('card',{'class':"block_header"})
len(cards)
```
