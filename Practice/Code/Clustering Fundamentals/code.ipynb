```python
k = 5
color = ["green","red","blue","yellow","orange"]

clusters = {}

for i in range(k):
    center = 10*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center': center,
        'points': points,
        'color':color[i]
    }
    clusters[i] = cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```
