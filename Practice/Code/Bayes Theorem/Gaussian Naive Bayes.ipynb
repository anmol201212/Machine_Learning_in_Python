## Multinomial Vs Gaussian Naive Bayes

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Data preparation

```python
# Load the Sklearn MNIST Dataset
digits = load_digits()
```

```python
X= digits.data
Y= digits.target
```

```python
X
```

```python
Y
```

```python
X.shape,Y.shape
```

```python
plt.imshow(X[0].reshape((8,8)))
print(Y[0])
plt.show()
```

```python
plt.imshow(X[4].reshape((8,8)))
print(Y[4])
plt.show()
```

### Train the Models

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB
```

```python
mnb = MultinomialNB()
gnb = GaussianNB()
```

```python
mnb.fit(X,Y)
gnb.fit(X,Y)
```

```python
print(mnb.score(X,Y))
print(gnb.score(X,Y))
```

```python
cross_val_score(mnb, X, Y,scoring="accuracy", cv=10).mean()
```

```python
cross_val_score(gnb, X, Y,scoring="accuracy",cv=10).mean()
```
