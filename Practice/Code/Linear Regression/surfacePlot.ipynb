## Surface Plots are used for
1. Visualise Loss Function in ML & DL
2. Visualise State or State Value Func in Reinforcement Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6,7])

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.add_subplot(111,projection='3d')
axes.plot_surface(a,b,a+b, cmap ='viridis')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(111,projection='3d')
axes.plot_surface(a,b,a**2+b**2, cmap ='rainbow')
plt.show()
```

```python
a = np.arange(-1,1,0.02)
b = a


a,b = np.meshgrid(a,b)
```

```python
fig = plt.figure()
axes = fig.add_subplot(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
```

```python
fig = plt.figure()
axes = fig.add_subplot(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
```
