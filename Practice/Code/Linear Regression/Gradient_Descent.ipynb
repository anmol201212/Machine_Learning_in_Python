# Gradient Descent

This is using derivative Method

Find a Local minima in any given function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

Given a function f(x), we want to find the value of x that minimizes f

```python
# Visualisation
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(x)")
plt.show()
```

```python
x = 0 
lr = 0.1
error = []
# 50 steps in the downhill direction
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
# Plot the values of error
plt.plot(error)
plt.show()
```
