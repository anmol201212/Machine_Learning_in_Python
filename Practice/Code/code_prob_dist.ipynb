```python
import numpy as np
from matplotlib import pyplot as plt
```

# Histogram


```python
# Generate a Normal Distribution (1-D)

u = 0
sigma = 1
# such a distribution is also called Standard Normal Distribution

vals = u + sigma*np.random.randn(1000)
print(vals.shape)

plt.hist(vals,50)
plt.show()
```

```python

u = 5
sigma = 2

vals = u + sigma*np.random.randn(1000)
print(vals.shape)

plt.hist(vals,10)
plt.show()
```

```python
vals = np.round(vals)
z = np.unique(vals,return_counts=True)
print(z)
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

# Multivariate Normal Distribution

```python
# Bivariate
mean = np.array([0.0,0.0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
```

```python
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]])

dist = np.random.multivariate_normal(mean,cov,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
```
